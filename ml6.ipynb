```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder

# Dummy dataset: Weight (grams), Texture (0 = Smooth, 1 = Rough)
X = np.array([[150, 0], [170, 1], [120, 0], [180, 1], [130, 0]])  

# Fruit labels
y = np.array(["Apple", "Orange", "Apple", "Melon", "Apple"])  

# Encode the target labels into numeric values
le = LabelEncoder()
y_encoded = le.fit_transform(y)  # Converts ["Apple", "Orange", "Melon"] to [0, 1, 2]

# Initialize Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y_encoded)

# Print decision tree rules
print("Decision Tree Classifier Rules:")

# Convert class names to a list to avoid the error
class_names_list = le.classes_.tolist()

# Visualize the Decision Tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=["Weight", "Texture"], class_names=class_names_list, filled=True)
plt.show()

# Predict fruit type for a new input
sample_input = np.array([[160, 1]])  # Example: 160g and rough texture
predicted_label = clf.predict(sample_input)
print("\nPredicted Fruit:", le.inverse_transform(predicted_label))


# Display the decision tree visualization
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=["Weight", "Texture"], class_names=le.classes_, filled=True)
plt.show()
```
